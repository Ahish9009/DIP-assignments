```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

# Loading Images
- Images are loaded from the *images* folder
- Images in OpenCV are loaded in the BGR color format, however MatPlotLib outputs in RGB. Hence, we convert from BGR to RGB using OpenCV's cvtColor method

```python
im1 = cv.imread("../images/im1.jpg")
im1 = cv.cvtColor(im1, cv.COLOR_BGR2RGB)

im2 = cv.imread("../images/im2.png")
im2 = cv.cvtColor(im2, cv.COLOR_BGR2RGB)

im3 = cv.imread("../images/im3.png")
im3 = cv.cvtColor(im3, cv.COLOR_BGR2RGB)

im4 = cv.imread("../images/im4.jpg")
im4 = cv.cvtColor(im4, cv.COLOR_BGR2RGB)
```

# Plotting a 2x2 Plot

- `This is done using MatPlotLib`

```python
fig = plt.figure(figsize=(15,15))

plt.subplot(221)
plt.imshow(im1)
plt.title("Image-1")

plt.subplot(222)
plt.imshow(im2)
plt.title("Image-2")

plt.subplot(223)
plt.imshow(im3)
plt.title("Image-3")

plt.subplot(224)
plt.imshow(im4)
plt.title("Image-4")
```

# Converting to Grayscale without using a Function

- The following formula is used for the conversion
    - `pixel_val = 0.3*R + 0.59*G + 0.11*B`
- The image is saved using MatPlotLib's `imsave` method


```python
im1_bw = np.zeros(im1.shape[:2])
im1_bw = 0.3*im1[:,:,0] + 0.59*im1[:,:,1] + 0.11*im1[:,:,2]

plt.imshow(im1_bw, cmap='gray')
plt.imsave("../images/im1_bw.png", im1_bw, cmap="gray")
```

```python
minv, maxv = np.min(im3), np.max(im3)
print("Range is:", minv, "to", maxv)
```
